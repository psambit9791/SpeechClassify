# Reading the files and Creating a Pandas dataframe

```python
from scipy.io import wavfile
import pandas as pd
import os

song_dict = {}
speech_type = {'rap':1, 'singing':2, 'speech':3}
for root, dirs, files in os.walk("./data/"):
    for file in files:
        if file.endswith('.wav'):
            song_id = os.path.join(root, file)
            fs, data = wavfile.read(song_id)
            song_dict[file.split('.')[0]] = (data, fs, speech_type[root.split('/')[-2]])
song_df = pd.DataFrame.from_dict(song_dict, orient='index', columns=['Data', 'Freq', 'Type'])
```

```python
song_df.to_hdf('song_df.h5', key='song_df')
```
